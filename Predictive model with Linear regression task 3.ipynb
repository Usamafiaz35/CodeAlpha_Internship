```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"D:\\Data Science\\cgpa_vs_package.csv")
data.head(3)
```

```python
data.isnull().sum()
```

```python
# decide the X (independent) & Y(dependent)
x = data[["CGPA"]]
y = data["Package"]
```

```python
# check data is Linear or not
plt.figure(figsize = (5,3))
sns.scatterplot(x="CGPA" , y="Package" , data=data)
plt.show()
```

```python
# splitting data into train and test 
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# calling the linear regression (line equation)
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train , y_train)
```

```python
# find accuracy of model
lr.score(x_test , y_test)*100
```

```python
2.75   5.59
3.46	7.01
```

```python
lr.predict([[3.46]])
```

```python
# x k corespond y ki kiya output aygi 
y_pre = lr.predict(x)
```

```python
# show the line of output where it passes

plt.figure(figsize = (5,3))
sns.scatterplot(x="CGPA" , y="Package" , data=data)
plt.plot(data["CGPA"] , y_pre , c="red")
plt.legend(["org data" , "predict line"])
# plt.savefig("predict.jpg")
plt.show()
```
